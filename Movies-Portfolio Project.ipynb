```python
# Import the libraries
import pandas as pd
import seaborn as sns
import numpy as np
import matplotlib
import matplotlib.pyplot as plt
plt.style.use('ggplot')
from matplotlib.pyplot import figure

%matplotlib inline
matplotlib.rcParams['figure.figsize'] = (12,8)

pd.options.mode.chained_assignment = None
```

```python
# Create the dataframe using the dataset

df= pd.read_csv(r"C:\Users\Admin\Desktop\movies.csv")
```

```python
# Checking the dataframe

df.head()
```

```python
df.columns
```

```python
# Checking for missing data
for col in df.columns:
  pct_missing = np.mean(df[col].isnull())
  print("{} = {}%".format(col, round(pct_missing*100)))
```

```python
# Dropping the null values

df = df.dropna()

#Checking for null values
for col in df.columns:
  pct_missing = np.mean(df[col].isnull())
  print("{} = {}%".format(col, round(pct_missing*100)))
```

```python
#Checking datatypes of columns

df.dtypes
```

```python
# Changing datatype of budget and gross

df['budget'] = df['budget'].astype('int64')
df['gross'] = df['gross'].astype('int64')
```

```python
df =df.sort_values(by=['gross'], inplace=False, ascending=False) 
```

```python
# Scatter plot between budget abd gros income
plt.scatter(x= df["budget"], y= df["gross"])
plt.title("Budget vs gross earnings")
plt.xlabel("Budget for films")
plt.ylabel("Gross Earnings")
plt.show()
```

```python
df.head()
```

```python
# Budget vs gross earnings using seaborn
sns.regplot(x= "budget", y= "gross", data = df,scatter_kws = {"color":"red"}, line_kws = {"color": "blue"} )
```

```python
# Creating correlation matrix

df.corr()
```

```python
correlation_matrix = df.corr()
sns.heatmap(correlation_matrix, annot = True)
plt.title("Correlation matrix for numeric values")
plt.xlabel("Movie Features")
plt.ylabel("Movie Features")
plt.show()
```

```python
df_numerized = df

for col_names in df_numerized.columns:
  if(df_numerized[col_names].dtype == 'object'):
        df_numerized[col_names]= df_numerized[col_names].astype('category')
        df_numerized[col_names]=df_numerized[col_names].cat.codes
            
df_numerized.head()
```

```python
correlation_matrix = df_numerized.corr()
sns.heatmap(correlation_matrix, annot = True)
plt.title("Correlation matrix for numeric values")
plt.xlabel("Movie Features")
plt.ylabel("Movie Features")
plt.show()
```

```python
df_numerized.corr()
```

```python
correlation_mat = df_numerized.corr()
corr_pairs = correlation_mat.unstack()

```

```python
sorted_pairs = corr_pairs.sort_values()
sorted_pairs
```
